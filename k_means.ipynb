```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = load_iris()
X = iris.data[:, :2] #first parameter ':' indicates that all rows of the dataset should be considered
#second parameter ':2' indicates that first two columns of the dataset should be considered
# ie,consider only the first 2 features of iris for X
#print(X)
#print(iris.data)
y = iris.target
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, init='k-means++', random_state=21)
km.fit(X)
centers = km.cluster_centers_
print(centers)
new_labels = km.labels_
```

```python
fig, axes = plt.subplots( 1,2, figsize=(16,8)) #subplot(a,b,c) a rows,b columns and c-th figure
axes[0].scatter(X[:, 0], X[:, 1], c=y)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='prism')
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[0].set_title('Actual', fontsize=18)
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[1].set_title('Predicted', fontsize=18)
```
